```python
class ContaBancaria:

# Atributos(dados)
  def __init__(self,titular, saldo):
    self.titular = titular
    self.saldo = saldo

# Métodos(comportamento)
  def depositar(self, valor):
    if valor <= 0:
      print("Valor invalido para deposito!!")
    else:
      self.saldo += valor
      print(f"Deposito efetuado com sucesso!! Você possui R${self.saldo:.2f}.")

  def sacar(self, valor):

    if valor <= 0:
      print("Valor invalido para saque!!")
    elif valor <= self.saldo:
      self.saldo -= valor
      print(f"Saque realizado com sucesso!!. Você possui R${self.saldo:.2f}.")
    else:
      print(f"Saldo insuficiente!!. Você possui R${self.saldo:.2f}.")

  def extrato(self):
    print(f"Titular: {self.titular}, Saldo atual: {self.saldo:.2f}")



```

```python
pessoa_1 = ContaBancaria("Eduardo", 2000.00)
```

```python
pessoa_1.depositar(2000)
```

```python
pessoa_1.sacar(2000)
```

```python
pessoa_1.extrato()
```
